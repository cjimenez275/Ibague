## Generador de datos

```python
import random
import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
genero = ["m", "f", "nr", "f", "f"]
```

```python
N = 100
L_genero = []
L_edad = []
L_horas = []
```

```python
for i in range(N):
    L_genero.append(random.choice(genero))
    x = abs(random.gauss(25, 5))
    y = abs(random.gauss(4, 0.8))
    L_edad.append(int(x))
    L_horas.append(y)
```

```python
hora = np.array(range(1, 10))
cone = lambda x: x**2 - 1
```

## Gráficas con un dato

```python
# Barras
cuentas = [L_genero.count("f"),
          L_genero.count("m"),
          L_genero.count("nr")]
plt.bar(["f", "m", "nr"], cuentas)
plt.show()
```

```python
# Histograma
plt.hist(L_edad)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.violinplot(L_horas)
plt.show()
```

```python
plt.eventplot(L_horas, orientation="vertical")
plt.show()
```

## Gráficas de dos datos

```python
plt.plot(hora, cone(hora))
plt.xlabel("Hora")
plt.ylabel("Conexión")
plt.show()
```

```python
x = np.linspace(1, 12)
f = lambda x: x**2
m = lambda x: x**3 - x**2
plt.plot(x, f(x), label="Femenino")
plt.plot(x, m(x), label="masculino")
plt.legend()
plt.xlabel("Hora")
plt.ylabel("Conexión")
plt.show()
```

```python
# Dispersión
plt.scatter(L_edad, L_horas)
plt.show()
```

```python
fig, ax = plt.subplots()
y = f(x)
ax.stackplot(x, y)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, f(x),'r')
ax2.plot(x, m(x), 'g')
plt.show()
```

## Gráficos en 3d

```python
fig, ax = plt.subplots()
x = L_edad
y = L_horas
ax.hist2d(x, y, bins=(20, 20), cmap=plt.cm.BuPu)
plt.xlabel("Edad")
plt.ylabel("Horas")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hexbin(x, y, gridsize=20, cmap=plt.cm.BuPu)
plt.show()
```

```python
x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)
X, Y = np.meshgrid(x, y)
f = lambda x,y: x**2 - y**2
Z = f(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
# make data with uneven sampling in x
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, ax = plt.subplots()

ax.pcolormesh(X, Y, Z, vmin=-0.5, vmax=1.0)

plt.show()
```
