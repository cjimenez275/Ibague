# Datos sobre la tensión en pacientes

```python
import pandas as pd
import random
import datetime
import matplotlib.pyplot as plt   
import seaborn as sns
```

## Generación de datos.
Se generan 670 datos de la edad y la tensión de pacientes para dos géneros

```python
# 16 a 24
Nm, Nf = 80, 60
sm, sf, dm, df = 122, 115, 74, 72
vsm, vsf, vdm, vdf = 12, 11, 12, 10
edad = []
genero = []
presions = []
presiond = []
fecha = []
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 12, 31)
dt = end_date - start_date
days_between_dates = dt.days
```

```python
for i in range(Nm):
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    fecha.append(random_date)
    edad.append("16 a 24")
    genero.append("masculino")
    presions.append(int(random.gauss(sm, vsm)))
    presiond.append(int(random.gauss(dm, vdm)))
for i in range(Nf):
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    fecha.append(random_date)
    edad.append("16 a 24")
    genero.append("femenino")
    presions.append(int(random.gauss(sf, vsf)))
    presiond.append(int(random.gauss(df, vdf)))
```

```python
# 25 a 39
Nm, Nf = 90, 70
sm, sf, dm, df = 127, 120, 78, 74
vsm, vsf, vdm, vdf = 14, 13, 10, 10
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 12, 31)
dt = end_date - start_date
days_between_dates = dt.days
```

```python
# 40 a 49
Nm, Nf = 100, 100
sm, sf, dm, df = 132, 130, 84, 78
vsm, vsf, vdm, vdf = 18, 20, 10, 11
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 12, 31)
dt = end_date - start_date
days_between_dates = dt.days
```

```python
# 40+
Nm, Nf = 90, 80
sm, sf, dm, df = 137, 136, 84, 85
vsm, vsf, vdm, vdf = 18, 19, 11, 12
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 12, 31)
dt = end_date - start_date
days_between_dates = dt.days
```

```python
for i in range(Nm):
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    fecha.append(random_date)
    edad.append("40+")
    genero.append("masculino")
    presions.append(int(random.gauss(sm, vsm)))
    presiond.append(int(random.gauss(dm, vdm)))
for i in range(Nf):
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    fecha.append(random_date)
    edad.append("40+")
    genero.append("femenino")
    presions.append(int(random.gauss(sf, vsf)))
    presiond.append(int(random.gauss(df, vdf)))
```
